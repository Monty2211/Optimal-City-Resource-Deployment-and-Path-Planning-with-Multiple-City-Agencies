```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
ours = 78,45,121,80,100,120,50,64,69,85,98,115,102,114,30,53,66,100,107,124,150,80,100,120,50,64,69,85,98,60,80,100,120,50,64,69,85,98,40,
tools = 90,60,119,88,103,132,53,66,89,80,120,120,115,126,54,73,89,93,114,150,158,100,123,122,53,79,85,80,114,72,90,123,140,96,66,70,108,118,70
stats.mannwhitneyu(ours, tools, alternative='two-sided')
```

```python
sum_ours=sum(ours)
sum_ours
```

```python
sum_tools=sum(tools)
sum_tools
```

```python
mean_ours=sum_ours/39
mean_ours
```

```python
mean_tools=sum_tools/39
mean_tools
```

```python
c_result = pd.DataFrame()
c_result['Total Distance'] = ours + tools
```

```python
result = pd.DataFrame()
result['OP-Route'] = ours
result['OR-Tools'] = tools
```

```python
plt.plot(result['OP-Route'], label='OP-Route')
plt.plot(result['OR-Tools'], label='OR-Tools')
plt.xlabel('Vehicle')
plt.ylabel('Distance')

plt.legend()
```

```python
for i in c_result.index[:39]:
    c_result['model'][i] = 'OP-Route'
for i in c_result.index[39:]:
    c_result['model'][i] = 'OR-Tools'
```
